```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(numeric_only='True'),annot=True,cmap='coolwarm',fmt='.2f')
```

```python
df.hist(figsize=(12, 8), bins=30)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df['carrier_delay'],df['arr_del15'],color = 'green')
plt.xlabel('Carrier delay')
plt.ylabel('Flight Delay more than 15 Minute')
plt.show()
```

```python
sns.pairplot(df[['arr_delay','carrier_delay','weather_delay']])
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df['carrier_ct'],df['arr_del15'],color = 'blue')
plt.xlabel('Carrier_ct',color='Green')
plt.ylabel('Arr_delay15',color='purple')
plt.title('Carrier delay due to carrier_ct',color='Red')
```

```python
delay_cols = [
    'carrier_delay',
    'weather_delay',
    'nas_delay',
    'security_delay',
    'late_aircraft_delay'
]
```

```python
total_delay = df[delay_cols].sum()
```

```python
plt.figure(figsize=(8,5))
total_delay.plot(kind='bar')

plt.title("Total Delay Minutes by Delay Reason")
plt.xlabel("Delay Reason")
plt.ylabel("Total Delay Minutes")

plt.xticks(rotation=45)
plt.show()
```

```python
percentage_delay = (total_delay / total_delay.sum()) * 100
```

```python
plt.figure(figsize=(7,7))
plt.pie(percentage_delay,labels=percentage_delay.index,autopct='%1.1f%%',startangle=90)
plt.title("Contribution of Delay Reasons to Total Arrival Delay")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['late_aircraft_delay'], df['arr_delay'], alpha=0.4)
plt.xlabel("Late Aircraft Delay (minutes)")
plt.ylabel("Arrival Delay (minutes)")
plt.title("Late Aircraft Delay vs Arrival Delay")
plt.show()
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent = missing_percent.sort_values(ascending=False)

print(missing_percent)
```

```python
missing = df.isnull().mean().sort_values(ascending=False) * 100
```

```python
plt.figure(figsize=(10,6))
missing[missing > 0].plot(kind='bar')
plt.ylabel('Missing Percentage (%)')
plt.title('Missing Values Percentage by Column')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title('Missing Values Heatmap')
plt.show()
```
